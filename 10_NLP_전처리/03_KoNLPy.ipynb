## 형태소 분석기 공통 메소드
- `morphs(string)` : 형태소 단위로 토큰화(tokenize)
- `nouns(string)` : 명사만 추출하여 토큰화(tokenize)    
- `pos(string)`: 품사 부착
    - 형태소 분석기 마다 사용하는 품사태그가 다르다.
        - https://konlpy-ko.readthedocs.io/ko/v0.5.2/morph/
- `tagset`: 형태소 분석기가 사용하는 품사태그 설명하는 속성.

# Konlpy와 nltk를 이용해 헌법 text 분석하기

```python
# 형태소 분석기 객체를 생성. 종류별로 클래스가 제공
from konlpy.tag import Okt, Kkma  #, Mecab은 라이브러리는 있지만 사용할 수 없음.

okt = Okt()  # 형태소 분석기 객체 생성 kkma = Kkma()
kkma = Kkma()

# 형태소 단위 토큰화
tokens_okt = okt.morphs(txt)
tokens_kkma = kkma.morphs(txt)

print(tokens_okt)
print(tokens_kkma)
```

```python
tokens_okt2 = okt.morphs(txt, stem=True)  # stem=True: 원형 복원 > Okt의 기능.
print(tokens_okt2)
```

```python
nouns = okt.nouns(txt)

print(nouns)
print(len(nouns))
```

```python
# 품사 부착
pos_tag = okt.pos(txt)
print(pos_tag)
print(okt.tagset)
```

```python
# OKT 기능 - 비속어 처리.
sample_txt = "이것도 해보던갘ㅋㅋㅋ."

okt.morphs(sample_txt, norm=True)  # norm(normalize)=True > 비속어 처리

# 비속어를 처리한 문장을 반환
okt.normalize(sample_txt)
```

```python
# 헌법 corpus를 분석
from nltk import Text, FreqDist
from konlpy.tag import Okt
from konlpy.corpus import kolaw
from wordcloud import WordCloud
```

```python
# data loading
with kolaw.open("constitution.txt") as f:
    txt = f.read()

print(txt)
```

```python
# 형태소 분석기 생성
okt = Okt()

# 전처리 + 토큰화
## 전처리: 명사, 동사 추출
pos_tag = okt.pos(txt)
print(pos_tag)

tokens = [token for token, pos in pos_tag if pos in ["Noun", "Verb"] and len(token) >= 2]  # 명사 or 동사만 추출
print(tokens[:5])
```

```python
# Text, FreqDist 분석 (EDA)
text = Text(tokens, name="헌법")  # Text()에는 token화된 데이터를 넘겨준다.
print(text)
```

```python
# 빈도 수 그래프
text.plot(30)
```

```python
print("법률: ", text.count("법률"))
print("대통령: ", text.count("대통령"))
```

```python
# 특정 단어들의 문서 내 분포
text.dispersion_plot(["법률", "대통령", "국가", "국회", "국민", "헌법", "자유"])
```

```python
freq = text.vocab()
print("총 단어 수: ", freq.N())
print("고유 단어 개수: ", freq.B())
```

```python
print("가장 많이 나온 토큰: ", freq.max())
print("'헌번'의 빈도 수, 비율: ", freq.get('헌법'), freq.freq('헌법'))
print(f"가장 많이 나온 단어와 빈도 수: {freq.max()}, {freq.get(freq.max())}")
```

```python
freq.most_common(10)
```

```python
wc = WordCloud(
    font_path="C:\\Windows\\Fonts\\malgun.ttf",  # 폰트 파일의 경로로
    max_words=200,
    min_font_size=2,
    relative_scaling=1,
    width=400,
    height=300,
    background_color="white",
    prefer_horizontal=0.5
)

wc_img = wc.generate_from_frequencies(freq)

wc_img.to_file("constitution_wc.png")
```
